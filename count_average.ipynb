```python
import pandas as pd

data = pd.read_csv('data_edited.csv')
```

```python
def count_avg(file):
    new = file[['price','amount']]
    return pd.Series({'unit_price': (new['price']/new['amount']),'total_price':(new['price']*new['amount'])})
```

```python
#print (count_avg(data))

data.apply(count_avg, axis = 1)
```
